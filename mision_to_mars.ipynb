```python
# visit the NASA Mars News site and scrape headlines
nasa_url = 'https://mars.nasa.gov/news/'
browser.visit(nasa_url)
#time.sleep(1)
nasa_html = browser.html
nasa_soup = BeautifulSoup(nasa_html, 'html.parser')

list_of_news = nasa_soup.find('ul', class_='item_list')
first_item_of_list = list_of_news.find('li', class_='slide')
header = first_item_of_list.find('div', class_='content_title').text
synoposis = first_item_of_list.find('div', class_='article_teaser_body').text

print(header)
print(synoposis)


#find_all('li', class_="result-row")
results1 = nasa_soup.find_all('li', class_="slide")
```

```python
results1
```

```python

# Loop through results to retrieve article title, header, and timestamp of article

for result in results1:
    #lists = soup.find_all('ul', class_ = 'item_list')
    #for list in lists:
        try:
        
        
   
            title = result.find('div', class_='content_title').text

            paragraph = result.find('div', class_='article_teaser_body').text

         #The time and date of article publication
            date = result.find('div', class_= 'list_date').text
    
        
            if(title and paragraph and date):
                print('-----------------')
                print(title)
                print(paragraph)
                print(date)
        
         # Dictionary to be inserted into MongoDB
                post = {
                    'title': title,
                    'article': paragraph,
                    'date': date
                }

# Insert dictionary into MongoDB as a document
                collection.insert_one(post)
        
        
        
        
        
        except AttributeError as e:
            print(e)
            

   

    
```
